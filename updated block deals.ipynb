```python
import pandas as pd
from urllib.request import Request, urlopen
from google.colab import drive
drive.mount('/content/drive')
url = 'https://www.tickertape.in/market-movers'
request = Request(url)
page = urlopen(request)
html_cont = page.read()
tables = pd.read_html(html_cont)
for table in tables:
  print(table, '/n/n')
for i, table in enumerate(tables,start=1):
  name = f'table_{i}.csv'
  new= table.to_csv('/content/drive/My Drive/INS2.xlsx',index=False)
df = pd.read_excel('/content/drive/My Drive/INSIDERF.xlsx')
df.drop(0, axis = 0, inplace = True)
```

```python
frames = [new, df]
result = pd.concat(frames, ignore_index=True)
result.to_excel('/content/drive/My Drive/INSIDERFinal.xlsx',index=False)
```
