```python
import pandas as pd
from dateutil import relativedelta
```

```python
df = pd.read_csv("/content/drive/MyDrive/base sheet for cost.csv")
df.columns= ['GA', 'Station','CMake','DMake', 'Location', 'Date']
df['Date']=pd.to_datetime(df.Date)
df.head()
```

```python
end  = input('Please give input for end date (in YYYY-MM-DD) and press enter - ')
```

```python

#creating new lists containing y, m, d of date column
years = []
months = []
days = []
for i in df['Date']:
  rd = relativedelta.relativedelta( pd.to_datetime(end), pd.to_datetime(i))
  y = rd.years
  years.append(y)
  m = rd.months
  months.append(m)
  d = rd.days
  days.append(d)
#creating new columns with y, m, d
df['Year'] = years
df['Months'] = months
df['Days'] = days
#creatingnew column with commisioning month number
df['Comminsioning Month'] = df['Date'].dt.month
df.head()
```

```python
#creating cost function
def cost_fn_cp(year, make):
    if year == 0:
      if make == 'SUN':
        return 0
      elif make == 'RAAS':
        return 0
      elif make == 'SOPAN':
        return 0
      elif make == 'ENDURANCE':
        return 29500
      elif make == 'IDEATE':
        return 0
      elif make == 'JYO':
        return 0
      elif make == 'KAL':
        return 0
      elif make == 'ICL':
        return 35400
      elif make == 'KPCL':
        return 64300
      elif make == 'DELTA':
        return 0
      elif make == 'DS':
        return 0
      elif make == 'CP':
        return 0
    if year == 1:
      if make == 'SUN':
        return 59000
      elif make == 'RAAS':
        return 53100
      elif make == 'SOPAN':
        return 70800
      elif make == 'ENDURANCE':
        return 56640
      elif make == 'IDEATE':
        return 47200
      elif make == 'JYO':
        return 51920
      elif make == 'KAL':
        return 53100
      elif make == 'ICL':
        return 0
      elif make == 'KPCL':
        return 76500
      elif make == 'DELTA':
        return 0
      elif make == 'DS':
        return 0
      elif make == 'CP':
        return 0
    if year == 2:
      if make == 'SUN':
        return 62540
      elif make == 'RAAS':
        return 56286
      elif make == 'SOPAN':
        return 75048
      elif make == 'ENDURANCE':
        return 60038.4
      elif make == 'IDEATE':
        return 50032
      elif make == 'JYO':
        return 55035.2
      elif make == 'KAL':
        return 56286
      elif make == 'ICL':
        return 0
      elif make == 'KPCL':
        return 0
      elif make == 'DELTA':
        return 0
      elif make == 'DS':
        return 0
      elif make == 'CP':
        return 0
    if year == 3:
      return 0
 #creating disp cost fn
def cost_fn_dp(year, make):
    if year == 0:
      if make == 'TGT':
        return 0
      elif make == 'PARK':
        return 11800
      elif make == 'COMP':
        return 0
    if year == 1:
      if make == 'TGT':
        return 5310
      elif make == 'PARK':
        return 12980
      elif make == 'COMP':
        return 0
    if year == 2:
      if make == 'TGT':
        return 6490
      elif make == 'PARK':
        return 14160
      elif make == 'COMP':
        return 0
    if year == 3:
      return 0
```

```python
li_days_in_prev_yr = []
cp = []
dp = []
a=0
for i in df['Months']:
  a+=1
  year = df['Year'][a-1]
  if i == 0:
#extracting commisioning month number for stations that are in transistion month (shifting from prev yr to next yr)
    date = df['Date'].loc[df.index[a-1]]
    p = pd.Period(date, freq='M')
#extracting number of days in that transition month
    total_days_in_month = p.days_in_month
#defining make
    CMake = df['CMake'][a-1]
    DMake = df['DMake'][a-1]
#calculating no. of days it ran in prev yr and no. of days in next year
    days_in_prev_yr = total_days_in_month - df['Days'].loc[df.index[a-1]]
    li_days_in_prev_yr.append(days_in_prev_yr)
    cal_cost_cp = days_in_prev_yr/total_days_in_month*cost_fn_cp(year-1, CMake) + df['Days'].loc[df.index[a-1]]/total_days_in_month*cost_fn_cp(year, CMake)
    cal_cost_dp = days_in_prev_yr/total_days_in_month*cost_fn_dp(year-1, DMake) + df['Days'].loc[df.index[a-1]]/total_days_in_month*cost_fn_dp(year, DMake)
    dp.append(cal_cost_dp)
    cp.append(cal_cost_cp)
    print(a, cal_cost_cp)
  else:
    days_in_prev_yr = 0
    CMake = df['CMake'][a-1]
    DMake = df['DMake'][a-1]
    li_days_in_prev_yr.append(days_in_prev_yr)
    cal_cost_cp = cost_fn_cp(year, CMake)
    cal_cost_dp = cost_fn_dp(year, DMake)
    dp.append(cal_cost_dp)
    cp.append(cal_cost_cp)
    print(a, cal_cost_cp)
print(cp,dp)
```

```python
df['Days in prev yr'] = li_days_in_prev_yr
df['CP_cost'] = cp
df['DP_cost'] = dp
```
